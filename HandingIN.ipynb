```python
#работаем с pixilbig.png -- картинка, состоящая из четырех строк пикселей

from PIL import Image
im = Image.open("Downloads/pixilbig.png")
import numpy as np
from PIL import Image    #  pip install Pillow

img = np.asarray(Image.open('Downloads/pixilbig.png').convert('RGB'))
#for row in img:
   # print (*row)
```

```python
#будем для каждого пикселя подсчитывать, сколько монотонных подпоследовательностей, 
#состоящих из пикселей в одном столбце, в них заканчиваются -- получим массив Result2
Result2 = [ [0]*img.shape[1] for i in range(img.shape[0]) ]
Fup={} #вспомогательный массив для неубывающих подпоследовательностей
Fdown={} #вспомогательный массив для невозрастающих подпоследовательностей
F={}
for k in range(0, img.shape[1]):
    for j in range(0,3): #проходим по всем трем компонентам цвета
        Fup[0]=1
        Fdown[0]=1
        F[0]=1
        for i in range(1,img.shape[0]):
            if (img[i,k][j]<img[i-1,k][j]):
                Fdown[i]=Fdown[i-1]+1
                Fup[i]=1
            if (img[i,k][j]>img[i-1,k][j]):
                Fup[i]=Fup[i-1]+1
                Fdown[i]=1
            if (img[i,k][j]==img[i-1,k][j]):
                Fdown[i]=Fdown[i-1]+1
                Fup[i]=Fup[i-1]+1
            if (j>0):
                F[i]=min(max(Fup[i], Fdown[i]), F[i])
            if (j==0):
                F[i]=max(Fup[i], Fdown[i])
    for i in range(0,img.shape[0]):
        Result2[i][k]=F[i]
#for i in range(img.shape[0]):
   # for k in range (img.shape[1]):
     #   print(Result2[i][k], end = ' ')
   # print()  
```
